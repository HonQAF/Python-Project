## Data Cleaning

```python
# Convert Funding and Valuation columns to integers
unicorn_companies['Valuation'] = unicorn_companies['Valuation'].astype(str).str.replace('[^\d\.]', '', regex=True)
unicorn_companies['Valuation'] = unicorn_companies['Valuation'].str.replace('B', '').astype(float)
```

```python
# Replace empty strings with NaN
unicorn_companies['Funding'].replace('', float('nan'), inplace=True)
```

```python
unicorn_companies.head()
```

```python
# Convert the column from string to float
unicorn_companies['Funding'] = pd.to_numeric(unicorn_companies['Funding'], errors='coerce')
```

## Question 1 - Which unicorn companies have had the biggest return on investment?

```python
# Calculate the ROI based on valuation and funding
unicorn_companies['Return_on_Investment'] = (unicorn_companies['Valuation'] - unicorn_companies['Funding']) / unicorn_companies['Funding']
```

```python
unicorn_companies.head()
```

```python
unicorn_companies.tail()
```

```python
company_with_biggest_roi = unicorn_companies['Return_on_Investment'].max()
company_with_biggest_roi
```

```python
# Filter out rows with infinite ROI
valid_roi_companies = unicorn_companies[unicorn_companies['Return_on_Investment'] != float('inf')]

# Find the index of the row with the highest non-infinite ROI
index_with_max_roi = valid_roi_companies['Return_on_Investment'].idxmax()

# Get the name of the company with the highest non-infinite ROI
company_name_with_max_roi = valid_roi_companies.loc[index_with_max_roi, 'Company']

print("Company with the biggest ROI:", company_name_with_max_roi)
```

## Question 2 - How long does it usually take for a company to become a unicorn? Has it always been this way?

```python
unicorn_companies['Year Joined'] = unicorn_companies['Date Joined'].str.split('-').str[0]
unicorn_companies['Month Joined'] = unicorn_companies['Date Joined'].str.split('-').str[1]
unicorn_companies['Day Joined'] = unicorn_companies['Date Joined'].str.split('-').str[2]

unicorn_companies
```

```python
# Convert 'Year Joined' and 'Year Founded' to datetime
unicorn_companies['Year Joined'] = pd.to_datetime(unicorn_companies['Year Joined'])
unicorn_companies['Year Founded'] = pd.to_datetime(unicorn_companies['Year Founded'], format='%Y')

# Calculate the time to become a unicorn and calculate the average time
unicorn_companies['Time to Unicorn'] = (unicorn_companies['Year Joined'].dt.year - unicorn_companies['Year Founded'].dt.year).astype(int)
average_time = unicorn_companies['Time to Unicorn'].mean()

print("Average time to become a unicorn (in years):", average_time)
```

## Question 3 - Which countries have the most unicorns? Are there any cities that appear to be industry hubs?

```python
# Count unicorns by country
country_counts = unicorn_companies['Country'].value_counts().head()

print("Top 5 Countries with the most unicorns:")
print(country_counts)
```

```python
# Count unicorns by city
city_counts = unicorn_companies['City'].value_counts().head()

print("Top 5 Cities with the most unicorns:")
print(city_counts)
```

## Question 4 - Which investors have funded the most unicorns?

```python
# Split and stack the 'Select Investors' column
investor_counts = unicorn_companies['Select Investors'].str.split(', ', expand=True).stack().value_counts().head()

print("Investors funding the most unicorns:")
print(investor_counts)
```

## Univariate Analysis

```python
# Univariate analysis: Distribution of valuation
plt.figure(figsize=(10, 6))
sns.histplot(data=unicorn_companies, x='Valuation', bins=20)
plt.xticks(rotation=90)
plt.xlabel('Valuation')
plt.ylabel('Frequency')
plt.title('Distribution of Valuation')
plt.show()
```

### Observation: From the chart above, the Majority of the Valuation are less than 25

## Bivariate Analysis

```python
# Bivariate analysis example: Valuation vs. Funding
plt.figure(figsize=(15, 25))
sns.scatterplot(data=unicorn_companies, x='Year Joined', y='Country')
plt.xticks(rotation=45)
plt.xlabel('Year Joined')
plt.ylabel('Country')
plt.title('Year Joined vs. Country')
plt.show()
```

### Observation: From the chart above, 2021 has the highest number of countries joined, While United States has companies join from 2011 - 2022

```python
plt.figure(figsize=(30, 8))
sns.boxplot(data=unicorn_companies, x='Country', y='Valuation')
plt.title('Box Plot of Valuation by Country')
plt.xticks(rotation=45)
plt.show()
```

## Observation: The chart shows United States has the highest total of valuation

## Multivariate Analysis

```python
# Valuation vs. Funding vs. Industry
plt.figure(figsize=(12, 20))
sns.scatterplot(data=unicorn_companies, x='Valuation', y='Funding', hue='Industry')
plt.title('Scatter Plot of Valuation vs. Funding by Industry')
plt.show()

```

```python
# Valuation vs. Funding vs. Country (Box Plot)
plt.figure(figsize=(30, 12))
sns.barplot(data=unicorn_companies, x='Country', y='Valuation', hue='Country')
plt.xticks(rotation=90)
plt.title('Box Plot of Valuation vs. Funding by Country')
plt.show()
```

### Observation: From the chart above, Sweden has the highest funding vs Bahamas with the with the highest valuation

```python
# Heatmap of Correlation Matrix
correlation_matrix = unicorn_companies.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
